```python
import yfinance as yf
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
```

```python
end_date = datetime.now()
kruz = pd.DataFrame(yf.download (tickers = "KRUZ", start = '2023-02-07', end = end_date.strftime('%Y-%m-%d'), interval = "1d"))
nanc = pd.DataFrame(yf.download (tickers = "NANC", start = '2023-02-07', end = end_date.strftime('%Y-%m-%d'), interval = "1d"))
s_and_p = pd.DataFrame(yf.download (tickers = "^GSPC", start = '2023-02-07', end = end_date.strftime('%Y-%m-%d'), interval = "1d"))
```

```python
kruz_change = pd.DataFrame(kruz['Close'].pct_change().dropna())
nanc_change = pd.DataFrame(nanc['Close'].pct_change().dropna())
s_and_p_change = pd.DataFrame(s_and_p['Close'].pct_change().dropna())
```

```python
plt.plot(kruz_change, label = 'KRUZ')
plt.plot(nanc_change, label = 'NANC')
plt.plot(s_and_p_change, label = 'S&P')
plt.legend()
plt.show()
```

```python
kruz_growth = ((kruz.iloc[-1]['Close'] - kruz.iloc[0]['Close'])/kruz.iloc[-1]['Close'])*100
nanc_growth = ((nanc.iloc[-1]['Close'] - nanc.iloc[0]['Close'])/nanc.iloc[-1]['Close'])*100
s_and_p_growth = ((s_and_p.iloc[-1]['Close'] - s_and_p.iloc[0]['Close'])/s_and_p.iloc[-1]['Close'])*100
print("Republican ETF Growth: %f" % (kruz_growth))
print("Democratic EFT Growth: %f" % (nanc_growth))
print("S&P 500 Growth: %f" % (s_and_p_growth))
```
